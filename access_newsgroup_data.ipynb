```python
import numpy as np
import pandas as pd
import os
import subprocess
```

```python
base_path = './20_newsgroups/'
```

```python
output = str(subprocess.check_output('ls {}'.format(base_path), shell=True))
subdirs = output[2:-1].split('\\n')
paths = [(base_path  + subdir + '/') for subdir in subdirs]

columns = [
    'newsgroups',
    'path',
    'from',
    'subject',
    'message-ID',
    'organization',
    'references',
    'date',
    'lines',
    'body'
]
```

```python
def get_field(input_string, field_name):
    """Takes raw newsgroup file (input_string) and field_name and extracts the
    text in question from the text file. Robust to non-standard formattings.
    (returns null value if the parse doesn't work.)
    """
    match = field_name[0].upper() + field_name[1:] + ':'
    try:
        return input_string.split(match)[1].split(u'\n')[0]
    except:
        print('could not match: ', match)
        return np.nan

def parse(input_string, newsgroups=None, columns=columns, index=0):
    """takes input_string (raw newsgroup file) and parses it, returning a pandas
    dataframe with the parsed data.
    """
    result = pd.DataFrame(columns=columns, index=[index])
    result.at[index, 'newsgroups'] = newsgroups
    for column in result.columns:
        if column not in ['newsgroups', 'body']:
            result.at[index, column] = get_field(input_string, column)
    lines = result.at[index, 'lines']
    try:
        result.at[index, 'body'] = input_string.split('Lines:')[1][len(str(lines))+2:]
    except:
        print('Could not find text body.')
    return result
```

```python
df = pd.DataFrame()
file = None
for path in paths:
    newsgroups = path.split('/')[2]
    for filename in os.listdir(path):
        filename = path + filename
        print(filename)
        file = open(filename)
        try:
            raw_text = file.read()
        except:
            continue
        data = parse(raw_text, newsgroups, index=len(df)) #Parse the contents of the file
        df = pd.concat([df, data], axis=0) #concatenate new data to dataframe
        file.close()
```
